### **CC57 - MACHINE LEARNING - GRAFICA DE DISPERSION Y REGRESION LINEAL**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Datos del ejemplo
horas_estudio = np.array([2, 3, 4, 5, 6]).reshape(-1, 1)
calificaciones = np.array([70, 75, 80, 85, 90])

# Crear el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(horas_estudio, calificaciones)

# Predecir los valores ajustados
calificaciones_pred = modelo.predict(horas_estudio)

# Datos para la gráfica
df = pd.DataFrame({
    'Horas de Estudio': horas_estudio.flatten(),
    'Calificación': calificaciones,
    'Calificación Predicha': calificaciones_pred
})

# Gráfica de dispersión y línea de regresión
plt.figure(figsize=(10, 6))
plt.scatter(df['Horas de Estudio'], df['Calificación'], color='blue', label='Datos Reales')
plt.plot(df['Horas de Estudio'], df['Calificación Predicha'], color='red', label='Línea de Regresión')
plt.title('Relación entre Horas de Estudio y Calificación')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación')
plt.legend()
plt.grid(True)

# Anotaciones
plt.annotate('Variable Dependiente (Y): Calificación', xy=(6, 90), xytext=(7, 85),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10, color='blue')
plt.annotate('Variable Independiente (X): Horas de Estudio', xy=(2, 70), xytext=(3, 65),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10, color='blue')
plt.annotate('Ecuación de la Recta: Y = 65 + 5X', xy=(4, 80), xytext=(5, 75),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10, color='red')
plt.annotate('Pendiente (β1): 5', xy=(5, 85), xytext=(6, 80),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10, color='red')
plt.annotate('Ordenada al Origen (β0): 65', xy=(0, 65), xytext=(1, 60),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10, color='red')

plt.show()
```
