### **CC57 - MACHINE LEARNING - VISUALIZACIONES E INTERPRETACION - MODELO REGRESION LINEAL**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Datos del ejemplo
horas_estudio = np.array([2, 3, 4, 5, 6]).reshape(-1, 1)
calificaciones = np.array([70, 75, 80, 85, 90])

# Crear el modelo de regresión lineal
modelo = LinearRegression()
modelo.fit(horas_estudio, calificaciones)

# Predecir los valores ajustados
calificaciones_pred = modelo.predict(horas_estudio)

# Calcular residuos
residuos = calificaciones - calificaciones_pred


```

###**Visualizaciones**

```python
# Gráfica de dispersión con línea de regresión
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(horas_estudio, calificaciones, color='blue', label='Datos Reales')
plt.plot(horas_estudio, calificaciones_pred, color='red', label='Línea de Regresión')
plt.title('Gráfica de Dispersión con Línea de Regresión')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación')
plt.legend()
plt.grid(True)

# Gráfica de residuos
plt.subplot(1, 2, 2)
plt.scatter(horas_estudio, residuos, color='purple', label='Residuos')
plt.axhline(y=0, color='red', linestyle='--')
plt.title('Gráfica de Residuos')
plt.xlabel('Horas de Estudio')
plt.ylabel('Residuos')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

#### **Resultados Esperados e Interpretación**

####**1. Gráfica de Dispersión con Línea de Regresión**

- Muestra los puntos de datos y cómo la línea de regresión se ajusta a estos puntos.

- Un buen ajuste se observa cuando la línea de regresión pasa cerca de la mayoría de los puntos.

#### **2. Gráfica de Residuos**

- Muestra los residuos (diferencias entre valores observados y predichos) en función de la variable independiente (horas de estudio).

- Idealmente, los residuos deberían estar dispersos aleatoriamente alrededor de la línea horizontal en cero.

- La ausencia de patrones claros indica que el modelo es adecuado.

- En el caso de los datos proporcionados, esperamos ver una distribución aleatoria de los residuos, confirmando que el modelo lineal es apropiado para estos datos.
