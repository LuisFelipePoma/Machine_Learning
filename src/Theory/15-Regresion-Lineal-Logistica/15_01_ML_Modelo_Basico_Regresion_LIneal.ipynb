### **CC57 - MACHINE LEARNING - GRAFICA DE DISPERSION Y REGRESION LINEAL**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Datos del ejemplo
horas_estudio = np.array([2, 3, 4, 5, 6]).reshape(-1, 1)
calificaciones = np.array([70, 75, 80, 85, 90])

# Crear el modelo de regresin lineal
modelo = LinearRegression()
modelo.fit(horas_estudio, calificaciones)

# Predecir los valores ajustados
calificaciones_pred = modelo.predict(horas_estudio)

# Datos para la grfica
df = pd.DataFrame({
    'Horas de Estudio': horas_estudio.flatten(),
    'Calificacin': calificaciones,
    'Calificacin Predicha': calificaciones_pred
})

# Grfica de dispersin y lnea de regresin
plt.figure(figsize=(10, 6))
plt.scatter(df['Horas de Estudio'], df['Calificacin'], color='blue', label='Datos Reales')
plt.plot(df['Horas de Estudio'], df['Calificacin Predicha'], color='red', label='Lnea de Regresin')
plt.title('Relacin entre Horas de Estudio y Calificacin')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificacin')
plt.legend()
plt.grid(True)

# Anotaciones
plt.annotate('Variable Dependiente (Y): Calificacin', xy=(6, 90), xytext=(7, 85),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10, color='blue')
plt.annotate('Variable Independiente (X): Horas de Estudio', xy=(2, 70), xytext=(3, 65),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10, color='blue')
plt.annotate('Ecuacin de la Recta: Y = 65 + 5X', xy=(4, 80), xytext=(5, 75),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10, color='red')
plt.annotate('Pendiente (1): 5', xy=(5, 85), xytext=(6, 80),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10, color='red')
plt.annotate('Ordenada al Origen (0): 65', xy=(0, 65), xytext=(1, 60),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10, color='red')

plt.show()
```
