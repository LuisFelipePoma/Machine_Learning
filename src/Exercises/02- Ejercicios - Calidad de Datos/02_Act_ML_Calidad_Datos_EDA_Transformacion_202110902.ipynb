### **3.4. Preprocesamiento de los datos**

```python
#Como solucionamos los datos faltantes? Eliminamos todos los datos faltantes?
#Utilice una tecnica para reemplazar los datos faltantes (una para valores numericos y otra para categoricos)
#2.Tratamiento 2
zomato_pre = zomato.copy()

numerics_columns = zomato.select_dtypes(include=[np.number]).columns[1:]
categorical_columns = zomato.select_dtypes(include=[np.object_]).columns
```

```python
zomato[numerics_columns].isnull().sum()
```

```python
zomato_pre[categorical_columns].isnull().sum()
```

```python
# Utilice una tecnica para reemplazar los datos faltantes para categoricos
zomato_pre["Cuisines"].fillna(zomato["Cuisines"].mode()[0], inplace=True)
```

```python
#Verificamos que no existan datos faltantes
zomato_pre[numerics_columns].isnull().sum()
```

```python
#Solucionar duplicidad en los datos (si existen)
zomato_pre.duplicated().sum()
```

```python
#Identificar variables con valores atipicos
plt.figure(figsize=(15, 10))
zomato_pre[numerics_columns].boxplot()
plt.xticks(rotation=90)
plt.show()
```

```python
def get_limits(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_limit = q1 - (1.5 * iqr)
    upper_limit = q3 + (1.5 * iqr)
    return lower_limit, upper_limit
```

```python
# De existir, eliminar Outliers describiendo que tecnica(s) ha utilizado en cada caso
# Encontrando RIC (rango intercuartilico)
avg_low, avg_upper = get_limits(zomato_pre, "Average Cost for two")

# Eliminar Outliers
zomato_pre = zomato_pre[
    (zomato_pre["Average Cost for two"] > avg_low)
    & (zomato_pre["Average Cost for two"] < avg_upper)
]

vote_low, vote_upper = get_limits(zomato_pre, "Votes")
zomato_pre = zomato_pre[(zomato_pre["Votes"] > vote_low) & (zomato_pre["Votes"] < vote_upper)]

zomato_pre[numerics_columns].boxplot()
```
