```python
#Estandarizamos las variables predictoras

rescalarX = StandardScaler().fit_transform(X)
dataX = pd.DataFrame(data=rescalarX, columns=X.columns)
dataX.head()
```

###**4.CLASIFICACION RANDOM FOREST**

#### **4.1. ENTRENAMIENTO**

```python
#Mantenemos el clasificador Naive Bayes y agregamos el de Random Forest

from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB #Naive Bayes
from sklearn.neighbors import KNeighborsClassifier #K-nn
```

```python
X_train, X_test, Y_train, Y_Test = train_test_split(dataX, Y,
                                                    random_state = 22, test_size=0.2)
```

```python
#Mantenemos el clasificador Naive Bayes y agregamos el de K-NN
models = []
models.append(("Naive Bayes", GaussianNB()))
models.append(("K-NN k=1", KNeighborsClassifier(n_neighbors=1))) #Instancia KNN
models.append(("K-NN k=3", KNeighborsClassifier(n_neighbors=3))) #Instancia KNN
models.append(("K-NN k=5", KNeighborsClassifier(n_neighbors=5))) #Instancia KNN
```

```python
models
```

```python
#Ejecutando la clasificacion
results = []
names =[]

for name, model in models:
  kfold = KFold(n_splits=10, random_state = 22, shuffle=True)
  cv_result = cross_val_score(model, X_train, Y_train, cv=kfold, scoring= 'accuracy')
  names.append(name)
  results.append(cv_result)

i = len(names)-1

for i in range(len(names)):
    print(names[i], results[i].mean())
```

#### **4.2. EVALUACION DEL MODELO**

```python
#Visualizacion para compara el modelo Naive Bayes con KNN

fig = plt.figure()
fig.suptitle("Comparacion Naive Bayes vs KNN")
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

```python
#Evaluacion de los resultados

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
#Modelo Naive Bayes
naive = GaussianNB()
naive.fit(X_train, Y_train)
naive_predictions = naive.predict(X_test)

#Modelo KNN k = 1
knn1 = KNeighborsClassifier(n_neighbors=1)
knn1.fit(X_train, Y_train)
knn1_predictions = knn1.predict(X_test)

#Modelo KNN k = 3
knn3 = KNeighborsClassifier(n_neighbors=3)
knn3.fit(X_train, Y_train)
knn3_predictions = knn3.predict(X_test)

#Modelo KNN k = 5
knn5 = KNeighborsClassifier(n_neighbors=5)
knn5.fit(X_train, Y_train)
knn5_predictions = knn5.predict(X_test)
```

```python
#Reportes de Clasificacion

print("Precision del test para Naive Bayes:")
print(accuracy_score(Y_Test, naive_predictions))
print(classification_report(Y_Test, naive_predictions))

print("Precision del test para 1-KNN:")
print(accuracy_score(Y_Test, knn1_predictions))
print(classification_report(Y_Test, knn1_predictions))

print("Precision del test para 3-KNN:")
print(accuracy_score(Y_Test, knn3_predictions))
print(classification_report(Y_Test, knn3_predictions))

print("Precision del test para 5-KNN:")
print(accuracy_score(Y_Test, knn5_predictions))
print(classification_report(Y_Test, knn5_predictions))
```

```python
#Visualizacion de los resultados a traves de Matriz de Confusion

conf_nb = confusion_matrix(Y_Test, naive_predictions)
conf_knn1 = confusion_matrix(Y_Test, knn1_predictions)
conf_knn3 = confusion_matrix(Y_Test, knn3_predictions)
conf_knn5 = confusion_matrix(Y_Test, knn5_predictions)
label = ["0","1"]
```

```python
#Matriz Confusion Naive Bayes
sns.heatmap(conf_nb, annot=True,
            xticklabels= label, yticklabels=label)

plt.title("Matriz de Confusion - Naive Bayes")
plt.show()
```

```python
#Matriz Confusion 1-KNN
sns.heatmap(conf_knn1, annot=True,
            xticklabels= label, yticklabels=label)

plt.title("Matriz de Confusion - KNN=1")
plt.show()
```

```python
#Matriz Confusion 3-KNN
sns.heatmap(conf_knn3, annot=True,
            xticklabels= label, yticklabels=label)

plt.title("Matriz de Confusion - KNN=3")
plt.show()
```

```python
#Matriz Confusion 5-KNN
sns.heatmap(conf_knn5, annot=True,
            xticklabels= label, yticklabels=label)

plt.title("Matriz de Confusion - KNN=5")
plt.show()
```
